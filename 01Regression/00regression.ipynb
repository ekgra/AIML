Regression models are used for predicting a real value.
- If independent variable is time, then you're predicting future values
- else you're predicting present but unknown values

Regression includes 
- Simple Linear Regression
- Multiple Linear Regression
- Polynomial Regression
- Support Vector Regression (SVR)
- Decision Tree Regression
- Random Forest Regression


# Evaluating Regression Model Performance

## R Squared

The greater the value of R squared the better.
- If R squared is 1 (max), then the model is a bit too much fitted to the data
- If R squared is 0.9, that's a very good model 



R squared has an issue with addition of new features (independent variables)
With addition of new independent variables, the SS<sub>tot</sub> does not change, while SS<sub>residual</sub> would reduce or stay the same. 
Therefore the R squared would only decrease with addition of new features

Hence, Adjusted R squared

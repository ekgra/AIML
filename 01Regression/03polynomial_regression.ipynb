## Polynomial Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
X
```

```python
y
```

## Linear fit - we will compare with Polynomial later

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(X, y)
```

```python
plt.scatter(X, y, color="red")
plt.plot(X, regression.predict(X), color="blue")
plt.title("Linear Regression result")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

## Polynomial fit

Play with the degree (2, 4) to see how it impacts the fit

```python
from sklearn.preprocessing import PolynomialFeatures
poly_regression = PolynomialFeatures(degree = 4)
X_poly = poly_regression.fit_transform(X)
lin_regression = LinearRegression()
y_poly = lin_regression.fit(X_poly, y)
```

```python
plt.scatter(X, y, color="red")
plt.plot(X, lin_regression.predict(poly_regression.fit_transform(X)), color="blue")
plt.title("Polynomial Regression result")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```
